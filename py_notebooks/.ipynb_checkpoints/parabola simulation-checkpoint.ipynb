```python
def get_parabola_coordinates(x_range):
    y_cords = []
    #equation y^2 = -4ax
    for x in x_range:
        y = np.sqrt(4*2*x) * ( -1 if x > 0 else 1 )
        y_cords.append(y)
    return y_cords    
```

```python
x_cords = range(-50,50)
y_cords  = get_parabola_coordinates(x_cords)
```

```python
plt.scatter(x_cords, y_cords)
plt.show()
```
