```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from moviepy.video.io.bindings import mplfig_to_npimage
import moviepy.editor as mpy
```

```python
duration = 20.
max_x = 20
max_y = 20
fig_mpl, ax = plt.subplots(1, figsize=(6, 5), facecolor='white')

ax.set_xlim([0,max_x])
ax.set_ylim([0,max_y])
def get_parabola_coordinates(max_x):
    x_cords = range(-max_x,max_x)
    y_cords = [(x**2+2*x+2) for x in x_cords ]
    #equation y^2 = -4ax
    #for x in x_range:
    #    y = np.sqrt(4*2*x) if x >= 0 else -1*np.sqrt(4*2*(-x))
    #    y_cords.append(y)
    return x_cords, y_cords    

def draw_elementary_pattern(ax, center):
    ax.add_artist(mpl.patches.Wedge(center, .2, 0, 360, width=.01))
#     ax.add_artist(mpl.patches.Circle(center, .2))

def make_frame_mpl(t):
    ax.cla()
    draw_elementary_pattern(ax,  (max_y-t)/duration)
    return mplfig_to_npimage(fig_mpl) # RGB image of the figure

animation = mpy.VideoClip(make_frame_mpl, duration=duration)
```

```python
animation.ipython_display(fps=20, loop=1, autoplay=1)
```

```python
animation.write_videofile("circles.webm", fps=20)
```
